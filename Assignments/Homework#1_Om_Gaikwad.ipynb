```python
import numpy as np
import matplotlib.pyplot as plt
```

## Question 1
Print the result of **A+B** and explain the mechanism behind this.

You answer:

```python
A = np.array([[0, 1, 2]])
B = np.array([[10, 10, 10], [20, 20, 20], [30, 30, 30]])
# Broadcast is essentially when Numpy performs operations on arrays that have different shapes
#In this case, A has a shape (1, 3) and B has shape (3, 3). Here, numpy modifies A to shape (3, 3) to make the operation possible
result = A + B

print ("A + B = ", result)
```

## Question 2
The broadcasting rule can be also applied to the multiplication. In this question, two numpy arrays of different shapes are given, check their shape first. And then, reshape one of the array so that you can produce exactly the same result given below.

```python
A = np.array([1, 2, 3, 4])
B = np.array([[1, 2, 3, 4]])

# Checking shapes
print ("Shape of A: ", A.shape)
print ("Shape of B: ", B.shape)

# to allow braodcsting, we will change the shape of A to (4, 1)
mod_A = A.reshape(4, 1)
# insert code
result = mod_A * B

print ("Multiplication result: \n", result)
```

## Question 3
Generate an array with 100 rows and 2 cols at random. Plot it by Matplotlib. You should consider the plotting setting in terms of titles, legends, names of the x and y axe, etc.

**Firstly, please set your student id as the seed number. Like np.random.seed(20016329).**

```python
np.random.seed(20002350) # Please use your student id as the seed number

# Generating the random data - 100 rows and 2 cols
array = np.random.rand(100, 2)

# Create the plot
fig, axs = plt.subplots(figsize=(8, 6))
axs.scatter(array[:, 0], array[:, 1], color='black', marker='o', label="Data Points")

axs.set_title('Scatter Plot of Randomly Generated Data from a uniform distribution over [0, 1)')
axs.set_xlabel('X Axis')
axs.set_ylabel('Y Axis')
axs.legend(loc='lower center', bbox_to_anchor=(0.5, -0.2))  # Legend at the bottom


# Show the plot
plt.show()
```

## Question 4
For the previous generated data, consider each row of the array as an individual data point, with the two columns representing its distinct features. Identify the data points with the smallest and largest sum of features. Mark them in red and blue in the figure respectively.

Hint: You can compute the sum of each row and store it in an array named "sums". Then find the minimum or maximum index of this array.

```python
# Compute the sum of each row
sums = np.sum(array, axis=1)

# Identify the indices of the min and max sum
min_index = np.argmin(sums)
max_index = np.argmax(sums)

print(f"Smallest sum index: {min_index}, Largest sum index: {max_index}")

# Create the plot
fig, axs = plt.subplots(figsize=(8, 6))
axs.scatter(array[:, 0], array[:, 1], color='black', marker='o', label="Data Points")

# Smallest sum point in red
axs.scatter(array[min_index, 0], array[min_index, 1], color='red', marker='o', label="Smallest Sum")

# Largest sum point in blue
axs.scatter(array[max_index, 0], array[max_index, 1], color='blue', marker='o', label="Largest Sum")

# Set plot
axs.set_title('Scatter Plot with Extremes Marked')
axs.set_xlabel('Feature 1')
axs.set_ylabel('Feature 2')
axs.legend(loc='lower center', bbox_to_anchor=(0.5, -0.2), ncol=3)  # Legend at the bottom


plt.show()

```

## Question 5
Add two lines representing the mean of the Feature 1 and the 25% quantile of the Feature 2 seperately. Add the legend of the two lines.

```python
# Compute the mean of Feature 1 and the 25% quantile of Feature 2
mean_feature1 = np.mean(array[:, 0])
quantile_25_feature2 = np.quantile(array[:, 1], 0.25)

# Create the plot
fig, axs = plt.subplots(figsize=(8, 6))
axs.scatter(array[:, 0], array[:, 1], color='black', marker='o', label="Data Points")

#Vertical line for the mean of Feature 1
axs.axvline(x=mean_feature1, color='green', linestyle='--', label="Mean of Feature 1")

#Horizontal line for the 25% quantile of Feature 2
axs.axhline(y=quantile_25_feature2, color='purple', linestyle='-.', label="25% Quantile of Feature 2")

# Set plot title
axs.set_title('Scatter Plot with Mean and 25% Quantile Lines')
axs.set_xlabel('Feature 1')
axs.set_ylabel('Feature 2')
axs.legend(loc='lower center', bbox_to_anchor=(0.5, -0.2), ncol=3)  # Legend at the bottom


plt.show()
```
